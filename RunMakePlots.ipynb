```python
def plot_var(list_var,samples,var_str,plt_str,path,directory):
    fig=plt.figure(figsize=(16,8))
    #for l in range(0,len(list_var)):
    plt.hist(list_var,label=samples,bins=100,stacked=True,alpha=0.4)

    plt.title(directory)
    plt.yscale('log')
    plt.xlabel(var_str)
    plt.legend()
    #plt.savefig(path+directory+'/'+var_str+'_'+plt_str+'.pdf')
    #plt.show()
    
```
